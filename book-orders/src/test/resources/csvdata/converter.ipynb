```python
import pandas as pd
import random
```

```python
df_customer = pd.read_csv('customer_data.csv', converters={'credit_card_number': str})
df_orderitems = pd.read_csv('order_items.csv', converters={'book_price': float})
df_payment = pd.read_csv('payment.csv', converters={'credit_card_number': str, 'amount': float})
df_order = pd.read_csv('book_order.csv', converters={'amount': float})
```

```python
creditcards = df_customer['credit_card_number'].unique()
cantons = {
    'Kanton Bern': 'BE',
    'Kanton Zürich': 'ZH',
    'Kanton Luzern': 'LU',
    'Kanton Basel-Stadt': 'BS',
    'Kanton Graubünden': 'GR'
}
```

```python
df_customer['credit_card_number'] = df_customer['credit_card_number'].apply(lambda x : creditcards[random.randint(1, creditcards.size-1)] if x == '' else x)
df_customer['address_state_province'] = df_customer['address_state_province'].apply(lambda x: cantons[x])
```

```python
df_payment['credit_card_number'] = df_payment['credit_card_number'].apply(lambda x : creditcards[random.randint(1, creditcards.size-1)] )
payment_ids = df_payment['transaction_id'].unique()
```

```python
def getTotalAmount(order_id):
    df_temp = df_orderitems.query("order_id == " + str(order_id))
    product = df_temp['book_price'] * df_temp['quantity']
    return float("{:.2f}".format(product.sum()))
```

```python
df_order['address_state_province'] = df_order['address_state_province'].apply(lambda x: cantons[x])
```

```python
df_order['payment_id'] = df_order['payment_id'].apply(
    lambda x : x if x in payment_ids else payment_ids[random.randint(1, payment_ids.size-1)] )

df_order = df_order.set_index('id')

for order_id in df_order.index:
    total = getTotalAmount(order_id);
    df_order.at[order_id, 'amount'] = total
    
df_order = df_order.reset_index()
```

```python
df_payment = df_payment.set_index('transaction_id');


for payment_id in df_payment.index.unique():
    total_amount = df_order.query('payment_id == ' + str(payment_id))['amount']
    df_payment.at[payment_id, 'amount'] = total_amount
    
df_payment = df_payment.reset_index();
```

```python
df_order['order_date'] = pd.to_datetime(df_order['order_date'])
```

```python
df_customer.to_csv("customer_final.csv", index=False)
df_order.to_csv("order_final.csv", index=False)
df_orderitems.to_csv("oderitems_final.csv", index=False)
df_payment.to_csv("payment_final.csv", index=False)
```
